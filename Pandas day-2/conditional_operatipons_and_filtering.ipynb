```python
import pandas as pd
import os
import numpy as np
```

```python
df = pd.read_csv(os.path.join('..', 'sales.csv'), encoding='latin1')      
df
```

```python
df.dtypes
```

```python
df['CustomerID'].dropna(inplace=True)
```

#### Conditional Selection/ Filtering

##### 1) Boolean Indexing:

```python
df[df['Quantity']>5000]                       # Selecting rows where 'Quantity' is greater than 5000
```

```python
df[df['CustomerID'] % 9 == 0]
```

##### 2) .loc access

```python
df.loc[df['Country'] != 'United Kingdom',   'Country']                    #df.loc[x_label, ylabel]
```

##### 3) isin

```python
df[df['Quantity'].isin([90,250,350])]
```

```python
df[~df['Country'].isin(['United Kingdom', 'France', 'Australia', 'Germany'])]           #not operator + isin
```

#### Conditional Assignment/Creation of New Columns:

```python
df.loc[df['Quantity']<3, 'Country'] = 'Pakistan'
```

```python
df.loc[df['Quantity']<3, ['Quantity', 'Country']] 
```

##### using where() of numpy

```python
df['Price'] = np.where(df['Quantity']>1000, 'HIGH', 'Low')            #Creates a 'Price' column: 'High' if 'Qty' > 1000, 'Low' otherwise                    
```

```python
df
```

```python
# Making changes in existing col  np.where(condition, new, existing)
df['Price'] = np.where(df['Price'] == 'Low', 'LOWLow', df['Price'])
df
```

#### Conditional Operations on Entire DataFrames/Series:

#### The df.where() method in Pandas is used to replace values in a DataFrame where a specified condition is false. It functions similarly to an "if-then-else"
* It evaluates a boolean condition for each element in the DataFrame. If the condition is True for an element, that element's original value is retained. If the condition is False, the element is replaced.

```python
data = {'A': [10, 20, 30, 40],
        'B': [5, 15, 25, 35]}
df_new = pd.DataFrame(data)
df_new
```

```python
# df_new_modified = df_new.where(df_new['A'] > 20)
```

```python
# df_new_modified
```

##### By default, values that do not meet the condition are replaced with NaN (Not a Number). However, a different value can be specified using the other parameter.

```python
df_mew_modified = df_new.where(df_new['A'] > 20.0, other=0.0, inplace= True)
```
