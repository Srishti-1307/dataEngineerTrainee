```python
import pandas as pd
```

A Multi-level Index, also called a Hierarchical Index, means your DataFrame has more than one column as an index.

```python
index = pd.MultiIndex.from_tuples([('India', 'Mumbai'), ('India', 'Delhi'), ('USA', 'California'), ('USA', 'Mexico')], names=['country', 'city'])
```

```python
df = pd.DataFrame({'population':[300,400,380,420]}, index=index)
df
```

Accessing data in multi-level index

```python
df.loc['India']
```

Reset index: Makes multi-level index back to columns

```python
df.reset_index()
```

Set index again

```python
df.set_index(['country', 'city'], inplace=True)
```

Creating multi-index from columns


```python
df = pd.DataFrame({
    'Country': ['India', 'India', 'USA', 'USA'],
    'City': ['Delhi', 'Mumbai', 'New York', 'Chicago'],
    'Population': [19000000, 20000000, 8500000, 2700000]
})
df
```

```python
df.set_index(['Country', 'City'])
```
