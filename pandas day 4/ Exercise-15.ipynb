#### Task
* Group data by region and category.
* Calculate total sales and average order value.
* Sort regions by total revenue.

```python
df = pd.read_csv('sales_data.csv')
df.head()
```

```python
df.groupby(['Country', 'Product_Category'])['Cost'].agg(['sum', 'mean'])
```

```python
df.groupby('Country')['Revenue'].sum().sort_values(ascending=False)
```
