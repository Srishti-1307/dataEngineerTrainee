 #### Use rolling averages to smooth out daily sales data over 7-day windows.

```python
df = pd.read_csv('sales_data.csv', nrows=100)
df.drop(columns= ['Customer_Age', 'Customer_Gender', 'Date','Sub_Category', 'Age_Group', 'Product', 'Order_Quantity', 'Unit_Cost'], inplace=True)
```

```python
df.sort_values(by = 'Day', inplace=True)
```

```python
df['rolling_average_revenue'] = df['Revenue'].rolling(window=7).mean()
```
