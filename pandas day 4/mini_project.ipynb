##### Project: Generate daily, weekly, monthly sales summaries

```python
df = pd.read_csv('sales_data.csv')
df.head()
```

```python
df['Product_Category'].unique()
```

```python
summary = df.pivot_table(index= ['Year', 'Month', 'Day'], columns= ['Product_Category'], values= ['Revenue', 'Profit'], aggfunc= 'sum')
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

summary.sort_index(level='Month')
```

```python
summary2= df.groupby(['Product_Category', 'Year', 'Month', 'Day'])['Revenue'].sum()
```
