```python
import pandas as pd
```

##### pivot table is a powerful data summarization tool that allows you to reorganize and aggregate data from a DataFrame in a spreadsheet-style format.

```python
data = {
    'Region': ['East', 'West', 'East', 'West', 'East'],
    'Product': ['A', 'B', 'A', 'C', 'B'],
    'Sales': [100, 150, 120, 200, 180]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

#### df.pivot_table() ---- structure
* df.pivot_table(index, columns, values, aggfunc, fill_value, margins, dropna)
* index- column(s) to use as the new row index of the pivot table
* columns- column headers of the pivot table.f
* values- column(s) whose values will be aggregated and displayed in the pivot table's cells.
* aggfunc- 'mean', 'sum', 'count', 'min', 'max', etc.
* fill_value- to replace missing (NaN) values
* dropna- If set to True (default), rows with NaN values in any column specified in index, columns, or values are dropped before aggregation.

```python
df.pivot_table(index='Region',  values='Sales', aggfunc=sum)
```

```python
df.pivot_table(index='Region',  values='Sales', columns='Product', aggfunc=sum)
```
