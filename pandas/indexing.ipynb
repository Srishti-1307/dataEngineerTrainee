#### Row Selection

##### loc

```python
df.loc[100, 'Quantity']                   #df.loc[row_label, col_label]
```

```python
df.loc[100]                            # row access by row_label
```

```python
df.loc[[99, 100, 101,105,106,107]]                 # multiple row access is also possible.
```

```python
df.loc[99:101]          #both inclusive
```

```python
df.loc[99:101, ['InvoiceNo', 'Description', 'Quantity']]              #df.loc[row_label(kaha se:kaha tk), col_label(list me)]
```

```python
df.loc[df['Quantity']>400]         #try .shape() with it
```

#### df.loc[] ka structure; 
* df.loc[row_indexer, column_indexer]
* row_indexer:
  * directly row_label ---> ('row_name')
  * list of row labels ---> (['row1', 'row3', 'row7'])
  * slicing ---->           ('start_row':'end_row')  both inclusive
  * A boolean array of the same length as the DataFrame's index "used for conditional row selection" (df['column_name'] > 10)
  ----------------------------------------------------------------------------------------------------------------------------------------

* col_header:
  * directly column label ('column_name')
  * list of column labels (['col_A', 'col_C'])
  * slicing ('start_col':'end_col')  both inclusive

#### iloc

```python
df.iloc[5]                #give row-index
```

```python
df.iloc[[2,4,5]]       #multiple row-indices you can give
```

```python
df.iloc[1:7]          #slicing based! start inclusive, end exclusive
```

```python
df.iloc[1:8, 2:4]         #df.iloc[row-index, col-index]
```

```python
df.iloc[99:105, [2,3]]
```

```python
df.iloc[[99,100,101], [2,3,4]]
```
