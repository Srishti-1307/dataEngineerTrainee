```python
df2 = pd.read_csv('data.csv')
df2.head()
```

##### Below I am setting a col of data.csv as index 

```python
df2.set_index('Id')
```

```python
pd.read_csv('data.csv', index_col=False)       
# index_col=False: This explicitly tells pandas not to use any column as the index 
# instead generate a default integer index (0, 1, 2, ...).
# This is particularly useful when the first column might appear to be an index but is actually part of the data
```

##### drop parameter in set_index(): If True, the specified column(s) used for the index will be removed from the DataFrame's columns.

```python
df2.set_index('Id', drop=False)           #by-default drop=True
```

```python
pd.read_csv('data.csv', skiprows = 1).head(5)               # you can also provide list (of row numbers(integer)) in skiprows
```

```python
pd.read_csv('data.csv', skipfooter = 3)           ## Skips the last 3 rows
```

#### Below I am reading df without col name

```python
pd.read_csv('data.csv', header=None, skiprows=1)    #used skiprows cuz col header 1st row ban ke aa ra tha
```

#### Below I want to give customized col names and do not include file's col headers

```python
pd.read_csv('data.csv', names=['A', 'B', 'C', 'D', 'E', 'F'], header=None, skiprows=1)
```
