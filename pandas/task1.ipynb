#### glob
* glob is a standard library module used for finding pathnames matching a specified pattern
* It is particularly useful for tasks involving file system navigation and file management
* such as retrieving a list of files with a specific extension or located within certain directories. 

#### Task: read multiple csv files from csv files folder into a single data frame

```python
all_files = glob.glob('/home/developer/pandas/multiple_files_folder'  + "/*.csv")    

# here glob.glob() finds all files ending with .csv within the specified path
```

```python
all_files
```

```python
l1=[]
```

```python
for i in all_files:
    df = pd.read_csv(i, encoding = 'latin1')
    l1.append(df)
```

```python
combined_df = pd.concat(l1, ignore_index = True)
```

##### pd.concat
* structure: pd.concat(list, axis, ignore_index)
* By default, original indexes are preserved.
* "ignore_index=True" parameter can be used to reset and create a new, contiguous index for the combined object.
* When concatenating objects with differing columns (for axis=0) or differing rows (for axis=1): 
   * pd.concat() fills missing values with NaN by default.
